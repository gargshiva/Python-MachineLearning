```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x=np.arange(1,11)
y1=x*2
y2=x*3
print(x)
print(y1)
print(y2)
```

```python
#Line Plot
plt.plot(x,y1,color='blue',linewidth='1')
plt.plot(x,y2,color='red',linewidth='1')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#Bar Graph

result = {"Bob":90,"Joe":66,"Shaw":81,"Rin":32.5,"Shiva":87} 
names = list(result.keys()) 
marks = list(result.values())

plt.bar(names,marks,color='blue')
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,marks,color='blue')
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Scatter Plot
x=[7,9,19,10,13,23]
y1=[3,5,16,12,9,21]
y2=[5,8,16,14,19,27]

plt.scatter(x,y1,color='blue')
plt.scatter(x,y2,color='red')
plt.grid(True)
plt.show()
```

```python
# India Literacy Analysis
goi = pd.read_csv("/Users/shigarg1/Downloads/GOI.csv")
goi = goi[(goi['Category'] == "State") & (goi['Literacy Rate (Persons) - Total - 2011'] > 85)]
goi.head(20)
```

```python
plt.bar(goi['Country/ States/ Union Territories Name'],goi['Literacy Rate (Persons) - Total - 2011'])
plt.ylabel("Literacy rates - 2011")
plt.xlabel("States")
plt.show()
```

```python
sns.boxplot(x=goi['Country/ States/ Union Territories Name'],y=goi['Literacy Rate (Persons) - Total - 2011'])
```

```python
plt.figure(figsize=(5,5))
plt.pie(goi['Literacy Rate (Persons) - Total - 2011'],labels=goi['Country/ States/ Union Territories Name'],autopct="%0.2f%%")
plt.show()
```
